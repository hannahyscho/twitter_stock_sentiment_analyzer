```python
import matplotlib.animation as animation 
```

```python
from matplotlib import style
import time
```

```python
style.use("ggplot")
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)

def animate(i):
    pullData = open("twitter-AAPL.txt", 'r').read()
    lines = pullData.split('\n')
    
    xar  = []
    yar = []
    
    x = 0
    y = 0
    
    for l in lines[-200:]:
        x += 1
        try:
            yar.append(float(l))
            xar.append(x)
        except:
            continue
    ax1.clear()
    ax1.plot(xar, yar)

    
ani = animation.FuncAnimation(fig, animate, interval=1000)
plt.show()
```
